Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
options(repr.plot.width = 4, repr.plot.height = 4, repr.plot.res= 500)

p = ggplot(penguins, aes(x=body_mass_g, y=sex))

p = p + geom_point(fill='black', color='black', alpha=0.5, size = 1, position = position_jitter(h = 0.15, w = 0))

p = p + facet_wrap(~species, ncol=1)

p = p + theme_minimal()

p = p + labs(x = 'Body mass in grams', y = ' ')

p = p + theme(strip.text = element_text(hjust = 0.1))

p = p + scale_x_continuous(label = scales::comma)

p
```
