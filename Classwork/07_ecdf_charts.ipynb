````r
ecdf = penguins %>%
    arrange(body_mass_g) %>%
    #^sorts from lowest to highest
    group_by(species) %>%
    #^groups them duh
    mutate(rn = row_number(), fraction_of_penguins = rn / max(rn))
    #^add a row called rn which would give a number and fraction would assign it's relationship in terms of percentage wise
    #assigns a position
ecdf %>% head
````

Using the `ecdf` table, create an **ECDF** chart of penguin body mass grouped by species.

````r
p = ggplot(ecdf, aes(x=body_mass_g, y=fraction_of_penguins, color=species))

p = p + geom_point(size=0.8)

p = p + geom_line()

p = p + theme_minimal()

p = p + scale_color_brewer(palette = 'Set2')

p = p + scale_x_continuous(labels = comma)
p = p + scale_y_continuous(labels = percent)

p = p + labs(x = 'Penguins body mass in grams', y = 'Percent of Penguins', color = 'Species', fill = 'Species')

p = p + theme(legend.position = 'top')

p
````
